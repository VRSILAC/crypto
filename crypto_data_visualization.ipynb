##Task 1 : Importing all the necessary libraries

```
import pandas as pd       #Data analysis and Manipulation
```

```
import plotly.offline as py           # create table
import plotly.graph_objs as go        #create candlestick charts             # Importing Plotly
import plotly.express as px           #create Violin plots, pie charts, box plot, scatter plots,line graph, area graph
```

```
import plotly.io as pio
pio.renderers.default = 'colab'     # To initialize plotly 
```

```
top_coins.groupby(['Currency']).count()
```

```
from plotly.figure_factory import create_table
```

```
table = create_table(top_coins.head(10), colorscale = 'plasma')
py.iplot(table)
```

##Task 3 : Data Visualization using Violin plots, Pie chart and Box plot

```
px.violin(top_coins, x = 'Currency', y = 'Volume')
```

```
px.pie(top_coins, values = 'Volume', names = 'Currency')
```

```
px.box(top_coins, x = 'Currency', y = 'Close')
```

```
px.box(top_coins, x = 'Currency', y = 'Close', log_y = True)
```

##Task 4: Data Visualization using Scatter plots

```
px.scatter(top_coins, x = 'Close', y = 'Volume', color = 'Currency', hover_data = ['High', 'Low', 'Date'], height = 400)
```

```
px.scatter(top_coins, x = 'Close', y = 'Volume', color = 'Currency', hover_data = ['High', 'Low', 'Date'], height = 400, log_x = True, log_y = True)
```

##Task 5: Crypto specific Data Visualization- Bitcoin using Line graph, Area graph and Candlestick charts

```
from google.colab import files
files.upload()
```

```
bitcoin = pd.read_csv('bitcoin_data.csv')
```

```
bitcoin
```

```
px.line(bitcoin, x = 'Date', y = 'Close')
```

```
px.area(bitcoin, x = 'Date', y = 'Volume')
```

```
fig = go.Figure(data = [go.Candlestick(x = bitcoin['Date'],
                                      open = bitcoin['Open'], 
                                      high = bitcoin['High'],
                                      low = bitcoin['Low'],
                                      close = bitcoin['Close']
                                       )])

fig.show()
```
